# Data-Driven Product Management: Conducting a Market Analysis

### **Introduction**

As a product manager for a fitness studio, understanding the current demand for digital fitness classes is crucial for strategic planning and growth. This analysis aims to:

1. **Identify Peak Global Interest:** Determine when global interest in the keyword 'workout' reached its highest point to gauge overall market demand.
2. **Compare Keyword Popularity Trends:** Assess how the popularity of key fitness-related keywords, such as 'home workout,' 'gym workout,' and 'home gym,' has shifted from the COVID-19 pandemic to the present.
3. **Evaluate Country-Specific Interest:** Find out which country among the United States, Australia, and Japan exhibits the highest interest in workouts, helping to identify key markets for expansion.
4. **Analyze Regional Interest in Home Workouts:** Compare the interest in home workouts between the Philippines and Malaysia to target marketing efforts and tailor digital fitness offerings to high-demand regions.

These insights will guide your fitness studio in aligning its digital fitness strategies with market demands, optimizing growth opportunities, and enhancing overall service offerings.

### **1. Peak Global Search Interest**

To identify when global interest in the keyword 'workout' was at its peak, we analyze the search trends data. This helps gauge the overall market demand for fitness-related content.

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df_workout = pd.read_csv("https://raw.githubusercontent.com/oliviagaffney/oliviagaffney-data-portfolio/main/Data-Driven%20Product%20Management%3A%20Conducting%20a%20Market%20Analysis/data/workout.csv")

# Convert 'month' column to datetime format
df_workout["month"] = pd.to_datetime(df_workout["month"], format='%Y-%m')

# Find the peak month for global 'workout' searches
peak_month = df_workout.loc[df_workout['workout_worldwide'].idxmax(), 'month']
year_str = peak_month.strftime('%Y')

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df_workout['month'], df_workout['workout_worldwide'], marker='o', linestyle='-', color='b')
plt.title('Global Search Interest for "Workout" Over Time')
plt.xlabel('Month')
plt.ylabel('Popularity Index')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"The global search for 'workout' was at its peak in the year: {year_str}")
```

### **2. Keyword Popularity Trends**

We will compare the popularity of 'home workout,' 'gym workout,' and 'home gym' during the COVID-19 pandemic and the current period to understand shifts in consumer preferences.

```python
# Load the data
df_three_keywords = pd.read_csv("https://raw.githubusercontent.com/oliviagaffney/oliviagaffney-data-portfolio/main/Data-Driven%20Product%20Management%3A%20Conducting%20a%20Market%20Analysis/data/three_keywords.csv")

# Convert 'month' column to datetime
df_three_keywords['month'] = pd.to_datetime(df_three_keywords['month'], format="%Y-%m")

# Define the COVID-19 period and current period
covid_start = "2020-01-01"
covid_end = "2021-12-31"
current_end = df_three_keywords["month"].max()

# Filter data for the COVID-19 period
covid_data = df_three_keywords[(df_three_keywords["month"] >= covid_start) & (df_three_keywords["month"] <= covid_end)]

# Filter data for the current period
current_data = df_three_keywords[df_three_keywords["month"] > covid_end]

# Calculate average popularity for each keyword during the COVID-19 period
covid_avg = covid_data[["home_workout_worldwide", "gym_workout_worldwide", "home_gym_worldwide"]].mean().round(2)

# Calculate average popularity for each keyword in the current period
current_avg = current_data[["home_workout_worldwide", "gym_workout_worldwide", "home_gym_worldwide"]].mean().round(2)

# Determine the most popular keyword during the COVID-19 period
peak_covid_keyword = covid_avg.idxmax()
peak_covid_value = covid_avg.max()

# Determine the most popular keyword in the current period
current_keyword = current_avg.idxmax()
current_value = current_avg.max()

# Save the answers as variables
peak_covid = peak_covid_keyword
current = current_keyword

# Print results
print(f"Most popular keyword during COVID-19 pandemic: {peak_covid} (Average popularity: {peak_covid_value})")
print(f"Most popular keyword now: {current} (Average popularity: {current_value})")

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(df_three_keywords['month'], df_three_keywords['home_workout_worldwide'], label='Home workout')
plt.plot(df_three_keywords['month'], df_three_keywords['gym_workout_worldwide'], label='Gym workout')
plt.plot(df_three_keywords['month'], df_three_keywords['home_gym_worldwide'], label='Home gym')
plt.xticks(rotation=90)
plt.legend()
plt.title('Popularity Trends of Fitness Keywords')
plt.xlabel('Month')
plt.ylabel('Popularity')
plt.show()
```

### **3. Country with Highest Interest in Workouts**

To identify which country among the United States, Australia, and Japan has the highest interest in workouts, we will analyze the interest levels for the keyword 'workout' from 2018 to 2023. This insight can help prioritize market entry and focus areas.

```python
# Load the data
df_workout_geo = pd.read_csv("https://raw.githubusercontent.com/oliviagaffney/oliviagaffney-data-portfolio/main/Data-Driven%20Product%20Management%3A%20Conducting%20a%20Market%20Analysis/data/workout_geo.csv")

# Define the countries of interest
countries_of_interest = ["United States", "Australia", "Japan"]

# Filter the DataFrame for the countries of interest
filtered_df = df_workout_geo[df_workout_geo["country"].isin(countries_of_interest)]

# Find the country with the highest 'workout' interest
highest_workout_interest = filtered_df.loc[filtered_df['workout_2018_2023'].idxmax()]

# Extract the top country
top_country = highest_workout_interest["country"]

print(f'The country with the highest interest in workouts is: {top_country}')
```

### **4. Interest in Home Workouts: Philippines vs. Malaysia**

To determine which of the two countriesPhilippines or Malaysiahas a higher interest in home workouts, we will analyze the data on the popularity of the 'home workout' keyword. This will help target marketing efforts and tailor offerings to regions with higher demand.

```python
# Load the data
df_three_keywords_geo = pd.read_csv("https://raw.githubusercontent.com/oliviagaffney/oliviagaffney-data-portfolio/main/Data-Driven%20Product%20Management%3A%20Conducting%20a%20Market%20Analysis/data/three_keywords_geo.csv")

# Rename column to match other DataFrames
df_three_keywords_geo.rename(columns={'Country': 'country'}, inplace=True)

# Define the countries of interest
countries_of_interest = ['Philippines', 'Malaysia']

# Filter the data for the selected countries
filtered_df = df_three_keywords_geo[df_three_keywords_geo['country'].isin(countries_of_interest)]

# Find the country with the highest interest in 'home workout'
max_interest_country = filtered_df.loc[filtered_df['home_workout_2018_2023'].idxmax(), 'country']

# Save the result in a variable
home_workout_geo = max_interest_country

print(f"The country with the highest interest in home workouts is: {home_workout_geo}")
```
